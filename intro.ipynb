# Value Imputaion

```python
# In scikit learn we can impute values using simpleimputer
```

```python
from sklearn.impute import SimpleImputer
```

```python
df.age.isnull().sum()
```

```python
# SimpleImputer has three imp arguments 

# first one is placeholder which is missing values
# second one is strategy which is mean median or most_frequent
# third one is fill_value which is used when strategy is constant
```

```python
imp= SimpleImputer(strategy='mean')
df['age']=imp.fit_transform(df[['age']])
```

```python
df.age.isnull().sum()
```

```python
import numpy as np
def get_parameters(df):
    parameters ={}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype == 'float64' or df[col].dtype == 'int64' or df[col].dtype =='int32':
            strategy = 'mean'
        else:
            strategy = 'most_frequent'
            
        missing_values = df[col].isnull().values[0]
        parameters[col] ={
            'missing_values': np.nan,
            'strategy': strategy
        }
    return parameters        
        
get_parameters(df)        
                  
```

```python
parameters = get_parameters(df)
parameters
```

```python
for col , param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']
    imp = SimpleImputer(missing_values=missing_values, strategy=strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```

```python
df.isnull().sum()
```
